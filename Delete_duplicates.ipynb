# Delete duplicate tweets keep only one amomg the duplicates.

    Sort the tweets based on id , so tweets with same id will be placed one after the other.
    Next using drop_duplicates, we drop/ delete rows which have same tweet id.
    Out of the duplicate tweets only the first tweet is ket rest are dropped 
    lastly 'inplace = True' means the chances made will be reflected in the dataframe or the duplicate rows are deleted
    permanently from the dataframe 'df'.

```python
df.sort_values("ID", inplace = True) 
df.drop_duplicates(subset ="ID", keep = 'first', inplace = True)
```

```python
df.info()
```

        after deleting we see the number of tweets / rows in the dataframme re reduced 
        this happens as the duplicate entries are dropped.
        
        after deleting the tweets the index might get changed ,they will contain missing values as rows are deleted.
        we need to reset the index so the the index maintains a serialorder n our dataset.

```python
df.head()
```

```python
df=df.reset_index(drop= True)
df.head()
```

```python
df.info()
```
